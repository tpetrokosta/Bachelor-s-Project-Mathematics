```python
import pandas as pd


file_a = "weeks_on_chart.xlsx"
df = pd.read_excel(file_a, sheet_name="Sheet1")


ae_values = df.loc[0:76, 'Pop'].dropna().astype(str)

ak_values = df.loc[0:15, 'Easy Listening'].dropna().astype(str)
aq_values= df.loc[0:187, 'Hip hop/ Rap'].dropna().astype(str)
ax_values= df.loc[0:49, 'Country'].dropna().astype(str)
bi_values= df.loc[0:22, 'Latin'].dropna().astype(str)
bn_values= df.loc[0:58, 'Electronic'].dropna().astype(str)
br_values= df.loc[0:39, 'Rock/ Alternative'].dropna().astype(str)
bu_values= df.loc[0:47, 'R&B/ Soul'].dropna().astype(str)

print(ae_values)
print(ak_values)
print(aq_values)
print(ax_values)
print(bi_values)
print(bn_values)
print(br_values)
print(bu_values)


ae_set = set(ae_values)
ak_set= set(ak_values)
aq_set=set(aq_values)
ax_set=set(ax_values)
bi_set=set(bi_values)
bn_set=set(bn_values)
br_set=set(br_values)
bu_set=set(bu_values)

# Function to assign genre
def get_genre(x):
    x = str(x)
    if x in ae_set:
        return "Pop"
    elif x in ak_set:
        return "Easy Listening"
    elif x in aq_set:
        return "Hip hop/ Rap"
    elif x in ax_set:
        return "Country"
    elif x in bi_set:
        return "Latin"
    elif x in bn_set:
        return "Electronic"
    elif x in br_set:
        return "Rock"
    elif x in bu_set:
        return "R&B"
    else:
        return ""


df['Genre'] = df['spotify_genre'].apply(get_genre)



df.to_excel("output.xlsx", index=False)
```
